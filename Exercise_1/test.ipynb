```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def electromagnetic_wave(t, frequency, amplitude=1, phase=0):
    """
    Generate an electromagnetic wave function.

    Parameters:
        t (array): Time array.
        frequency (float): Frequency of the wave.
        amplitude (float, optional): Amplitude of the wave. Default is 1.
        phase (float, optional): Phase of the wave in radians. Default is 0.

    Returns:
        array: Electromagnetic wave function.
    """
    omega = 2 * np.pi * frequency
    return amplitude * np.sin(omega * t + phase)

# Parameters
t = np.linspace(0, 2, 1000)  # Time array from 0 to 2 seconds
frequency = 2  # Frequency of the wave (in Hz)
amplitude = 1.5  # Amplitude of the wave
phase = np.pi / 4  # Phase of the wave (45 degrees in radians)

# Generate electromagnetic wave
wave = electromagnetic_wave(t, frequency, amplitude, phase)

# Calculate magnetic field (orthogonal to electric field)
magnetic_field = np.gradient(wave)

# Scaling factors for electric and magnetic fields
electric_field_scaling = 0.2  # Decrease electric field magnitude
magnetic_field_scaling = 10.0  # Increase magnetic field magnitude

# Plot the wave in 3D with electric and magnetic fields
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot electromagnetic wave
#ax.plot(t, np.zeros_like(t), wave, color='blue', label='Electromagnetic Wave')

# Plot electric field (along x-axis)
ax.quiver(t, np.zeros_like(t), np.zeros_like(t), np.zeros_like(t), wave * electric_field_scaling, np.zeros_like(t),
          color='red', length=0.1, arrow_length_ratio=0.4, alpha = 0.1, label='Electric Field')

# Plot magnetic field (along y-axis)
ax.quiver(t, np.zeros_like(t), np.zeros_like(t), np.zeros_like(t), np.zeros_like(t), magnetic_field * magnetic_field_scaling,
          color='green', length=0.1, arrow_length_ratio = 0.4, alpha = 0.2, label='Magnetic Field')

ax.set_title('Electromagnetic Wave propagating in time')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Electric Field')
ax.set_zlabel('Magnetic Field')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the incident angle (in degrees) and refractive indices
incident_angle_deg = 30
n1 = 1.0  # Refractive index of the first medium (usually air)
n2 = 1.5  # Refractive index of the second medium (e.g., glass)

# Convert angle to radians
theta1 = np.deg2rad(incident_angle_deg)

# Calculate the angle of refraction using Snell's Law
theta2 = np.arcsin(n1 * np.sin(theta1) / n2)

# Plotting
plt.figure(figsize=(8, 6))

# Incident ray
plt.plot([0, np.cos(theta1)], [0, np.sin(theta1)], 'b', label='Incident Ray')

# Reflected ray
plt.plot([0, -np.cos(theta1)], [0, np.sin(theta1)], 'r', label='Reflected Ray')

# Transmitted ray
plt.plot([0, np.cos(theta2)], [0, -np.sin(theta2)], 'g', label='Transmitted Ray')

# Normal line
plt.plot([0, 0], [-1, 1], 'k--', label='Normal')

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Snell's Law Illustration")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
