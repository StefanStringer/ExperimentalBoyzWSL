```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 100
```

```python
theta1 = np.linspace(0, 90, 90)
theta2 = np.pi/2
```

```python
Rp = lambda theta: np.tan(theta - theta2)**2/(np.tan(theta + theta2)**2)
Rs = lambda theta: np.sin(theta - theta2)**2/(np.sin(theta + theta2)**2)
plt.plot(theta1, Rp(theta1), 'o')
plt.plot(theta1, Rs(theta1), 'o')
```
