```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
```

```python
intensity = np.array([2.399, 1.979, 1.964, 1.827, 2.299, 2.655, 2.714, 2.64, 2.504, 2.048, 1.786, 1.73, 1.739, 1.792, 2.115, 2.347, 2.557, 2.739, 2.684, 3.41, 3.479, 3.496, 3.365, 3.215, 2.952, 2.661, 2.446, 2.133, 2.062, 1.768, 2.011, 2.021, 2.161, 2.38, 2.541, 2.762, 3.02])
angles = np.arange(0, 370, 10)
```

```python
plt.plot(angles, intensity, 'o')
```

```python
coefficients, cov = np.polyfit(angles, intensity, 5, cov = True)  # Tryed with polyfit at first. Helps to determine how the degree should be and if the data is valid
x = np.linspace(0, 360, 1000)
y = np.polyval(coefficients, x)
plt.plot(x, y)
plt.plot(angles, intensity, 'o')
```
